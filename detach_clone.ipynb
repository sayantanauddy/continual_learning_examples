```python
import torch
%matplotlib inline
from torchviz import make_dot
```

```python
ip = torch.ones(1, 2, requires_grad=False)

weights = torch.rand(2, 1, requires_grad=True)
bias = torch.zeros(1, 1, requires_grad=True)

op = torch.mm(ip, weights) + bias
```

```python
make_dot(op)
```

```python
print(weights.grad)
print(bias.grad)
```

```python
op.backward()
```

```python
print(weights.grad)
print(bias.grad)
```

```python
print(weights)
print(bias)
```

```python
w_clone = weights.clone()
print(w_clone.requires_grad)

w_clone_detach = weights.clone().detach()
print(w_clone_detach.requires_grad)
```
